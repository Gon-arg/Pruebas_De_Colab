```python
import folium
import requests

# Límites de visión sobre Argentina
min_lon, max_lon = -74, -53
min_lat, max_lat = -55, -21

# Crear mapa centrado en Argentina con límites
m = folium.Map(
    location=[-38, -64],
    zoom_start=4,
    min_lat=min_lat,
    max_lat=max_lat,
    min_lon=min_lon,
    max_lon=max_lon,
    max_bounds=True,
)

# Descargar GeoJSON oficial de provincias de datos.gob.ar
url = "https://infra.datos.gob.ar/georef/provincias.geojson"
geojson = requests.get(url).json()  # ✅ contiene límites provinciales :contentReference[oaicite:1]{index=1}

# Dibujar provincias con color y bordes visibles
folium.GeoJson(
    geojson,
    style_function=lambda feat: {
        'fillColor': '#ADD8E6',
        'color': 'blue',
        'weight': 2,
        'fillOpacity': 0.2
    },
    tooltip=folium.GeoJsonTooltip(fields=['nombre'], aliases=['Provincia:'])
).add_to(m)

# Marcar esquinas del país
folium.CircleMarker([max_lat, min_lon], tooltip="Esquina Noroeste").add_to(m)
folium.CircleMarker([min_lat, min_lon], tooltip="Esquina Suroeste").add_to(m)
folium.CircleMarker([min_lat, max_lon], tooltip="Esquina Sureste").add_to(m)
folium.CircleMarker([max_lat, max_lon], tooltip="Esquina Noreste").add_to(m)

# Mostrar mapa
m
```

```python
import pandas as pd
import folium
from branca.colormap import linear
from IPython.display import display

# 0) Cargar dataset
df = pd.read_csv('/content/ConectividadalserviciodeInternet_2776151656793719256.csv')

# 1) Definir columnas de servicio
service_cols = ['ADSL', 'CABLEMODEM', 'DIALUP', 'FIBRAOPTICA',
                'SATELITAL', 'WIRELESS', 'TELEFONIAFIJA', '3G', '4G']

# 2) Cobertura total vs parcial
df['cobertura_total'] = df[service_cols].eq('SI').all(axis=1)
full_cov = df[df['cobertura_total']]
partial_cov = df[~df['cobertura_total']]

print("1) Localidades con cobertura TOTAL (todos los servicios = SI):")
display(full_cov[['Provincia', 'Partido', 'Localidad']])

print("\n2) Localidades SIN cobertura total (al menos un servicio NO disponible):")
display(partial_cov[['Provincia', 'Partido', 'Localidad'] + service_cols])

# 3) Cobertura promedio por provincia
df['servicios_disponibles'] = df[service_cols].eq('SI').sum(axis=1)
df['ratio_cobertura'] = df['servicios_disponibles'] / len(service_cols) * 100
prov_avg = df.groupby('Provincia')['ratio_cobertura'].mean().sort_values(ascending=False)

print("\n3) Top 3 provincias con mayor cobertura promedio (% servicios disponibles):")
display(prov_avg.head(3).round(2).rename("Cobertura media"))

print("\n4) Provincia con menor cobertura promedio:")
display(prov_avg.tail(1).round(2).rename("Cobertura media"))

# 5) Top 5 partidos en Buenos Aires por población y cobertura promedio
ba = df[df['Provincia'].str.upper() == 'BUENOS AIRES']
partidos = ba.groupby('Partido').agg(
    poblacion_total=('Poblacion', 'sum'),
    cobertura_media=('ratio_cobertura', 'mean')
).sort_values('poblacion_total', ascending=False)

print("\n5) En Buenos Aires, 5 partidos con mayor población y su cobertura (%):")
display(partidos.head(5).round(2))

# 6) Ciudades con población = 0
sin_pob = df[df['Poblacion'] == 0]
print("\n6) Ciudades con población 0:")
display(sin_pob[['Provincia', 'Partido', 'Localidad']])

# 7) Mapa: coordenadas de todas las ciudades, colores por provincia
provincias = df['Provincia'].unique()
colormap = linear.Set1_09.scale(0, len(provincias))
color_dict = {prov: colormap(i) for i, prov in enumerate(provincias)}

m = folium.Map(location=[-38, -64], zoom_start=4, min_zoom=4)

for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row['Latitud'], row['Longitud']],
        radius=3,
        color=color_dict[row['Provincia']],
        fill=True,
        fill_opacity=0.7,
        popup=(
            f"{row['Localidad']}<br>"
            f"{row['Provincia']} - {row['Partido']}<br>"
            f"Población: {row['Poblacion']}<br>"
            f"Cobertura: {row['ratio_cobertura']:.1f}%"
        )
    ).add_to(m)

print("\n7) Mapa interactivo de todas las ciudades (coloreadas por provincia):")
m
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Cargar datos (Iris dataset)
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

# Seleccionar características y target
X = df[['sepal length (cm)']].values  # Característica (longitud del sépalo)
y = df['target'].values  # Etiqueta (especies de flores)

# División de los datos en entrenamiento (70%) y prueba (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Modelo de regresión lineal
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)  # Entrenar el modelo
y_train_pred_linear = linear_model.predict(X_train)  # Predicciones en entrenamiento
y_test_pred_linear = linear_model.predict(X_test)  # Predicciones en prueba

# Modelo de regresión polinómica (grado 2)
poly_features = PolynomialFeatures(degree=2)
X_train_poly = poly_features.fit_transform(X_train)  # Transformar los datos
X_test_poly = poly_features.transform(X_test)
poly_model = LinearRegression()
poly_model.fit(X_train_poly, y_train)  # Entrenar el modelo polinómico
y_train_pred_poly = poly_model.predict(X_train_poly)  # Predicciones en entrenamiento
y_test_pred_poly = poly_model.predict(X_test_poly)  # Predicciones en prueba

# Métricas de evaluación (Error Cuadrático Medio - MSE)
mse_train_linear = mean_squared_error(y_train, y_train_pred_linear)
mse_test_linear = mean_squared_error(y_test, y_test_pred_linear)
mse_train_poly = mean_squared_error(y_train, y_train_pred_poly)
mse_test_poly = mean_squared_error(y_test, y_test_pred_poly)

# Resultados de los errores
print("MSE Entrenamiento - Regresión Lineal:", mse_train_linear)
print("MSE Prueba - Regresión Lineal:", mse_test_linear)
print("MSE Entrenamiento - Regresión Polinómica:", mse_train_poly)
print("MSE Prueba - Regresión Polinómica:", mse_test_poly)

# Visualización de los modelos
plt.figure(figsize=(14, 7))

# Gráfico de Regresión Lineal
plt.subplot(1, 2, 1)
plt.scatter(X, y, color='gray')  # Datos originales
plt.plot(X, linear_model.predict(X), color='red')  # Predicciones del modelo lineal
plt.title('Regresión Lineal')

# Gráfico de Regresión Polinómica
plt.subplot(1, 2, 2)
X_range = np.arange(X.min(), X.max(), 0.1).reshape(-1, 1)
X_range_poly = poly_features.transform(X_range)
plt.scatter(X, y, color='gray')  # Datos originales
plt.plot(X_range, poly_model.predict(X_range_poly), color='blue')  # Predicciones del modelo polinómico
plt.title('Regresión Polinómica (Grado 2)')

plt.show()
```
